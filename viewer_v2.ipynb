```python
import math
import pandas as pd

INTERVAL_STEP = 100
```

```python
def funcao_1(df_aux, column_group = 'G6', time_column = 'tempo_total', result_column = 'resultado'
             , interval = 100, agg_type = 'min', result_type = None):
    if result_type == True:
        df = df_aux[[column_group, time_column]].copy().loc[df_aux[result_column] == True]
    if result_type == False:
        df = df_aux[[column_group, time_column]].copy().loc[df_aux[result_column] == False]
    else:
        df = df_aux[[column_group, time_column]].copy()
    
    if df.empty == True:
        return df
    
    if agg_type == 'min':
        df = df.groupby([column_group]).min()
    elif agg_type == 'max':
        df = df.groupby([column_group]).max()
    else:
        df = df.groupby([column_group]).mean()

    max_aux = df[time_column].max()
    min_aux = df[time_column].min()
    mean_aux = df[time_column].mean()

    step = (max_aux - min_aux) / interval
    
    df['step'] = step
    
    return df, min_aux, mean_aux, max_aux
```
