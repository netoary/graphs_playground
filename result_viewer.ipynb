```python
import math
import pandas as pd

INTERVAL_STEP = 100
```

```python
def cluster_type(value):
    for i in range(INTERVAL_STEP):
        if value < (i + 1) * step:
            return i


def cluster_type2(args):
    value = args[-1]
    step = args[-2]
    for i in range(INTERVAL_STEP):
        if value < (i + 1) * step:
            return i
    return i+1
```

```python
def funcao_1(df_aux, column_group = 'G6', time_column = 'tempo_total', result_column = 'resultado'
             , interval = 100, agg_type = 'min', result_type = None):
    if result_type == True:
        df = df_aux[[column_group, time_column]].copy().loc[df_aux[result_column] == True]
    if result_type == False:
        df = df_aux[[column_group, time_column]].copy().loc[df_aux[result_column] == False]
    else:
        df = df_aux[[column_group, time_column]].copy()
    
    if df.empty == True:
        return df
    
    if agg_type == 'min':
        df = df.groupby([column_group]).min()
    elif agg_type == 'max':
        df = df.groupby([column_group]).max()
    else:
        df = df.groupby([column_group]).mean()

    max_aux = df[time_column].max()
    min_aux = df[time_column].min()
    mean_aux = df[time_column].mean()

    step = (max_aux - min_aux) / interval
    
    df['step'] = step
    df['cluster'] = df[[time_column, 'step']].apply(cluster_type2, axis=1)
    
    return df
```

```python
df = df_brute_test_12_all.head(1)
```

```python
df_min = funcao_1(df_brute_test_12_all, agg_type='min', result_type=True)
df_max = funcao_1(df_brute_test_12_all, agg_type='max', result_type=True)
df_mean = funcao_1(df_brute_test_12_all, agg_type='mean', result_type=True)

df_min['class'] = 'min'
df_max['class'] = 'max'
df_mean['class'] = 'mean'

df_all = pd.concat([df_min, df_max])
df_all = pd.concat([df_all, df_mean])
```
